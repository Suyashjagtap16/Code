```python
# ---- PART A: Data Wrangling on Iris ----

import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Check for missing values
print(df.isnull().sum())

# Introduce & Handle missing values
df.iloc[0, 0] = None
df.fillna(df.mean(), inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Filter data (Sepal length > 5)
filtered_df = df[df['sepal length (cm)'] > 5]
print("Filtered Data:\n", filtered_df.head())

# Normalize
df_normalized = (df - df.min()) / (df.max() - df.min())
print("Normalized Data:\n", df_normalized.head())


# ---- PART B: PySpark SQL on Book Dataset ----

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Book Dataset Analysis") \
    .getOrCreate()

data = [
    (1, "Book A", "Author X", 200),
    (2, "Book B", "Author Y", 150),
    (3, "Book C", "Author X", 300),
    (4, "Book D", "Author Z", 250)
]

columns = ["book_id", "title", "author", "price"]
df = spark.createDataFrame(data, columns)
df.show()

df.createOrReplaceTempView("books")

# Books with price > 200
result = spark.sql("SELECT * FROM books WHERE price > 200")
result.show()

# Average price
avg_price = spark.sql("SELECT AVG(price) AS avg_price FROM books")
avg_price.show()

spark.stop()
```
