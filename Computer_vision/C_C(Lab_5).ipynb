```python
# Step 1: Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Load Image

from google.colab import files
uploaded = files.upload()

image = cv2.imread('sunset.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Original Image")
plt.show()
```

```python
#Step 3: Pseudocolor Transformation
gray = cv2.imread('sunset.jpg', 0)
pseudo = cv2.applyColorMap(gray, cv2.COLORMAP_JET)
plt.imshow(pseudo)
plt.title("Pseudocolor Image")
plt.show()
```

```python
#Step 4: Full-Color Enhancement (CLAHE using LAB)
lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
l, a, b = cv2.split(lab)
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl = clahe.apply(l)
limg = cv2.merge((cl, a, b))
enhanced_img = cv2.cvtColor(limg, cv2.COLOR_LAB2RGB)
plt.imshow(enhanced_img)
plt.title("Enhanced Image (CLAHE)")
plt.show()
```

```python
#Step 5: Color-Based Segmentation (Example: Red Color Detection)
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])
mask = cv2.inRange(hsv, lower_red, upper_red)
result = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(result)
plt.title("Color Segmentation (Red)")
plt.show()
```
