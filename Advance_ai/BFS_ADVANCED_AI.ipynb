```python
# BFS

import numpy as np
from collections import deque

# Graph represented as adjacency matrix
nodes = ['A', 'B', 'C', 'D', 'E']

graph = np.array([
    [0,1,1,0,0],  # A
    [0,0,0,1,0],  # B
    [0,0,0,0,1],  # C
    [0,0,0,0,0],  # D
    [0,0,0,0,0]   # E
])

def bfs(start, goal):
    queue = deque([start])
    visited = set()

    while queue:
        node = queue.popleft()
        print("Visiting:", nodes[node])

        if node == goal:
            return "Goal Found!"

        visited.add(node)

        for i in range(len(graph[node])):
            if graph[node][i] == 1 and i not in visited:
                queue.append(i)

print(bfs(0, 4))  # A to E
```
