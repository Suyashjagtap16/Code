```python
print ("Name - Suyash Subhash Jagtap")
print ("PRN - 20250804077 ")

# Import Libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# ===== IMAGE ACQUISITION =====
print("Upload an image file")
uploaded = files.upload()

# Get uploaded file name
img_path = list(uploaded.keys())[0]

# Read image
img = cv2.imread(img_path)

# Check if image loaded
if img is None:
    print("Error: Image not loaded")
else:
    print("Image successfully loaded")

    # ===== DISPLAY ORIGINAL IMAGE =====
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis('off')
    plt.show()

    # ===== PIXEL INFORMATION =====
    print("Image Shape (Height, Width, Channels):", img.shape)
    print("Pixel value at (0,0):", img[0, 0])

    # ===== GRAYSCALE CONVERSION =====
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    plt.imshow(gray, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')
    plt.show()

    # ===== SAVE IMAGE IN DIFFERENT FORMATS =====
    cv2.imwrite('sample.png', img)
    cv2.imwrite('sample.bmp', img)

    print("Image saved in PNG and BMP formats")

    # ===== SAMPLING =====

    # Downsampling (reduce size)
    downsampled = cv2.resize(img_rgb, (200, 200))

    plt.imshow(downsampled)
    plt.title("Downsampled Image")
    plt.axis('off')
    plt.show()

    # Upsampling (increase size)
    upsampled = cv2.resize(img_rgb, (600, 600))

    plt.imshow(upsampled)
    plt.title("Upsampled Image")
    plt.axis('off')
    plt.show()
```
