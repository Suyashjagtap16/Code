```python
from pyspark.sql import SparkSession

# 1. Create Spark Session
spark = SparkSession.builder \
    .appName("RDD Operations") \
    .getOrCreate()
sc = spark.sparkContext

# 2. Create RDD
data = [10, 20, 30, 40, 50, 60]
rdd = sc.parallelize(data)
print("Original RDD:")
print(rdd.collect())

# 3. Transformations

# 3.1 Map (Multiply each element by 2)
mapped_rdd = rdd.map(lambda x: x * 2)

# 3.2 Filter (Keep values > 30)
filtered_rdd = rdd.filter(lambda x: x > 30)

# 3.3 FlatMap (Split words)
words = ["big data", "apache spark"]
word_rdd = sc.parallelize(words)
flat_rdd = word_rdd.flatMap(lambda x: x.split(" "))

# 3.4 Distinct
dup_data = [1, 2, 2, 3, 4, 4, 5]
dup_rdd = sc.parallelize(dup_data)
distinct_rdd = dup_rdd.distinct()

# 4. Actions

# 4.1 Collect
print("Mapped RDD:", mapped_rdd.collect())

# 4.2 Count
print("Count:", rdd.count())

# 4.3 First Element
print("First Element:", rdd.first())

# 4.4 Reduce (Sum of elements)
sum_value = rdd.reduce(lambda a, b: a + b)
print("Sum:", sum_value)

# 4.5 Take (First 3 elements)
print("Take 3:", rdd.take(3))

# 5. Additional Example (Word Count)
text = ["hello spark", "hello world", "big data spark"]
text_rdd = sc.parallelize(text)
word_count = text_rdd \
    .flatMap(lambda x: x.split(" ")) \
    .map(lambda word: (word, 1)) \
    .reduceByKey(lambda a, b: a + b)
print("Word Count:")
print(word_count.collect())

# 6. Stop Spark Session
spark.stop()
```
