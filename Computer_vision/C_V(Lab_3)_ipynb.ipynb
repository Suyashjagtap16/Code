```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

print("Mame - Suyash Subash Jagtap")
print("prn - 20250804077 ")

# Get uploaded filename
filename = list(uploaded.keys())[0]
print("Uploaded file:", filename)

# Read image
img = cv2.imread(filename)

# Check if image loaded
if img is None:
    print("Error: Could not load image")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply Gaussian blurs
    blur_15 = cv2.GaussianBlur(img_rgb, (15, 15), 5)
    blur_25 = cv2.GaussianBlur(img_rgb, (25, 25), 10)
    blur_45 = cv2.GaussianBlur(img_rgb, (45, 45), 15)

    # Display results
    plt.figure(figsize=(12, 10))

    plt.subplot(2, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original")
    plt.axis("off")

    plt.subplot(2, 2, 2)
    plt.imshow(blur_15)
    plt.title("Gaussian Blur (15x15, sigma=5)")
    plt.axis("off")

    plt.subplot(2, 2, 3)
    plt.imshow(blur_25)
    plt.title("Gaussian Blur (25x25, sigma=10)")
    plt.axis("off")

    plt.subplot(2, 2, 4)
    plt.imshow(blur_45)
    plt.title("Gaussian Blur (45x45, sigma=15)")
    plt.axis("off")

    plt.show()
```
