```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
audio = '/content/drive/MyDrive/audio1.mp3'
audio_np_array, sample_rate = librosa.load(audio, sr=None)

# Plotting the waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio_np_array, sr=sample_rate)
# Use waveshow instead of waveplot
plt.title('Audio Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
amplitude_envelope = np.abs(librosa.effects.preemphasis(audio_np_array))
librosa.display.waveshow(amplitude_envelope, sr=sample_rate)
plt.figure(figsize=(10, 4))
plt.plot(amplitude_envelope)
plt.title('Amplitude Envelope')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

```python
#loudness
loudness_db = librosa.amplitude_to_db(amplitude_envelope, ref=np.max)
librosa.display.waveshow(loudness_db, sr=sample_rate)
# Use waveshow instead of waveplot
plt.title('loudness')
plt.show()
```

```python
#Mel-frequency cepstral coefficients
mfccs = librosa.feature.mfcc(y=audio_np_array, sr=sample_rate, n_mfcc=13)
#n_mfcc = 13, Number of MFCCs to return
librosa.display.specshow(mfccs, sr=sample_rate)
plt.colorbar()
plt.title('MFCC')
plt.show()
```

```python
#chroma feature
chroma = librosa.feature.chroma_stft(y=audio_np_array, sr=sample_rate)
librosa.display.specshow(chroma, sr=sample_rate)
plt.colorbar()
plt.title('chroma waveform')
plt.show()
```

```python
# Step 3: Extract Spectral Contrast
spectral_contrast = librosa.feature.spectral_contrast(y=audio_np_array, sr=sample_rate)
librosa.display.specshow(spectral_contrast, sr=sample_rate)
plt.colorbar()
plt.title('spectral contrast')
plt.show()
```

```python
pitch_sal = librosa.piptrack(y=audio_np_array, sr=sample_rate)
pitch_values=np.argmax(pitch_sal, axis=1)
librosa.display.specshow(pitch_values, sr=sample_rate)
plt.colorbar()
plt.title('pitch saliency')
plt.show()
```

```python
# Calculate the spectral centroid
S=librosa.feature.melspectrogram(y=audio_np_array, sr=sample_rate)
spectral_centroid = librosa.feature.spectral_centroid(S=S)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), y_axis = 'mel', x_axis= 'time')
plt.semilogy(spectral_centroid, label='Spectral Centroid', color='b')
plt.title('Spectral Centroid')

plt.show()
```
