```python
#practice code 1
import string

def transform_text(text):
  text = text.lower()
  text = text.translate(str.maketrans('', '', string.punctuation))
  text = " ".join(text.split())
  tokens = text.split()
  vocab = {word: idx for idx, word in enumerate(tokens)}
  numbers = [vocab[token] for token in tokens]
  return tokens, numbers

tokens , numbers = transform_text("DSA , is a peice of CAKE!")
print(tokens)
print(numbers)
```

```python
import cv2
from matplotlib import pyplot as plt
from google.colab import files
import numpy as np

#upload image from computer
uploaded = files.upload() # this will open a file chooser

#get the uploaded file name
filename = list(uploaded.keys())[0]

#read the image using cv2
image = cv2.imread(filename)



#convert BGR to RGB for correct colors
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Display original image
plt.subplot(1, 2, 1) # 1 row, 2 columns, first subplot
plt.imshow(image)
plt.title('Original Image (BGR)')
plt.axis('off') # Hide axes for cleaner look

# Display RGB converted image
plt.subplot(1, 2, 2) # 1 row, 2 columns, second subplot
plt.imshow(image_rgb)
plt.title('Image after BGR to RGB Conversion')
plt.axis('off') # Hide axes for cleaner look

plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()
```
