```python
# Install required libraries (run once)
# !pip install nltk textblob

# Import libraries
import nltk
from textblob import TextBlob

# Download only required datasets (safe for Colab)
nltk.download('wordnet')
nltk.download('stopwords')

from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Initialize tools
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
tokenizer = TreebankWordTokenizer()

# Dataset (Marketing-related text)
data = [
    "I absolutely love this product, it has transformed my life!",
    "The new design is okay, but I expected something more innovative.",
    "I had a horrible experience with the customer service.",
    "This brand really understands what customers want. Highly recommended!",
    "The product is decent but nothing special.",
    "I'm never buying from this company again, totally disappointed."
]

# -----------------------------
# Preprocessing Function
# -----------------------------
def preprocess(text):
    tokens = tokenizer.tokenize(text.lower())  # Tokenization

    filtered_tokens = []
    for word in tokens:
        if word.isalnum() and word not in stop_words:  # Remove stopwords
            lemma = lemmatizer.lemmatize(word)         # Lemmatization
            filtered_tokens.append(lemma)

    return " ".join(filtered_tokens)

# -----------------------------
# Sentiment Analysis Function
# -----------------------------
def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  # -1 to +1

    if polarity > 0:
        sentiment = "Positive 😊"
    elif polarity < 0:
        sentiment = "Negative 😞"
    else:
        sentiment = "Neutral 😐"

    return polarity, sentiment

# -----------------------------
# Run Analysis
# -----------------------------
print("\n========= Sentiment Analysis Results =========\n")

positive_count = 0
negative_count = 0
neutral_count = 0

for i, text in enumerate(data):
    cleaned_text = preprocess(text)
    polarity, sentiment = analyze_sentiment(cleaned_text)

    # Count sentiments
    if "Positive" in sentiment:
        positive_count += 1
    elif "Negative" in sentiment:
        negative_count += 1
    else:
        neutral_count += 1

    print(f"Sentence {i+1}: {text}")
    print(f"Preprocessed: {cleaned_text}")
    print(f"Polarity Score: {round(polarity, 3)}")
    print(f"Sentiment: {sentiment}")
    print("-" * 60)

# -----------------------------
# Summary
# -----------------------------
print("\n========= Summary =========")
print(f"Positive: {positive_count}")
print(f"Negative: {negative_count}")
print(f"Neutral : {neutral_count}")
```
