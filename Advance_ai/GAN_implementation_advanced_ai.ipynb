```python
# ==========================================================
# GAN IMPLEMENTATION USING PYTORCH (CIFAR10 DATASET)
# ==========================================================

# -----------------------------
# 1. IMPORT LIBRARIES
# -----------------------------

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np

# Select GPU if available otherwise CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')


# -----------------------------
# 2. IMAGE PREPROCESSING
# -----------------------------

# Convert images to tensor and normalize pixel values
transform = transforms.Compose([
    transforms.ToTensor(),                         # Convert image to tensor
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5)) # Normalize to [-1,1]
])


# -----------------------------
# 3. LOAD DATASET
# -----------------------------

# Load CIFAR10 dataset
train_dataset = datasets.CIFAR10(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

# Create dataloader for batch processing
dataloader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True
)


# -----------------------------
# 4. HYPERPARAMETERS
# -----------------------------

latent_dim = 100      # Size of random noise vector
lr = 0.0002           # Learning rate
beta1 = 0.5           # Adam optimizer parameter
beta2 = 0.999
num_epochs = 10      # Number of training epochs


# -----------------------------
# 5. GENERATOR NETWORK
# -----------------------------

class Generator(nn.Module):

    def __init__(self, latent_dim):
        super(Generator, self).__init__()

        # Sequential neural network model
        self.model = nn.Sequential(

            # Convert noise vector into feature map
            nn.Linear(latent_dim, 128*8*8),
            nn.ReLU(),

            # Convert vector to 3D feature map
            nn.Unflatten(1,(128,8,8)),

            # Upsample image from 8x8 → 16x16
            nn.Upsample(scale_factor=2),

            # Convolution layer
            nn.Conv2d(128,128,kernel_size=3,padding=1),
            nn.BatchNorm2d(128,momentum=0.78),
            nn.ReLU(),

            # Upsample image from 16x16 → 32x32
            nn.Upsample(scale_factor=2),

            nn.Conv2d(128,64,kernel_size=3,padding=1),
            nn.BatchNorm2d(64,momentum=0.78),
            nn.ReLU(),

            # Final image output layer
            nn.Conv2d(64,3,kernel_size=3,padding=1),

            # Output pixel values between -1 and 1
            nn.Tanh()
        )

    # Forward propagation
    def forward(self,z):
        img = self.model(z)
        return img


# -----------------------------
# 6. DISCRIMINATOR NETWORK
# -----------------------------

class Discriminator(nn.Module):

    def __init__(self):
        super(Discriminator,self).__init__()

        self.model = nn.Sequential(

            # Convolution layer
            nn.Conv2d(3,32,kernel_size=3,stride=2,padding=1),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.25),

            nn.Conv2d(32,64,kernel_size=3,stride=2,padding=1),
            nn.ZeroPad2d((0,1,0,1)),
            nn.BatchNorm2d(64,momentum=0.82),
            nn.LeakyReLU(0.25),
            nn.Dropout(0.25),

            nn.Conv2d(64,128,kernel_size=3,stride=2,padding=1),
            nn.BatchNorm2d(128,momentum=0.82),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.25),

            nn.Conv2d(128,256,kernel_size=3,stride=1,padding=1),
            nn.BatchNorm2d(256,momentum=0.8),
            nn.LeakyReLU(0.25),
            nn.Dropout(0.25),

            # Flatten feature maps
            nn.Flatten(),

            # Final classification layer
            nn.Linear(256*5*5,1),

            # Output probability
            nn.Sigmoid()
        )

    def forward(self,img):
        validity = self.model(img)
        return validity


# -----------------------------
# 7. MODEL INITIALIZATION
# -----------------------------

generator = Generator(latent_dim).to(device)
discriminator = Discriminator().to(device)

# Binary Cross Entropy Loss
adversarial_loss = nn.BCELoss()

# Optimizers
optimizer_G = optim.Adam(generator.parameters(), lr=lr, betas=(beta1,beta2))
optimizer_D = optim.Adam(discriminator.parameters(), lr=lr, betas=(beta1,beta2))


# -----------------------------
# 8. TRAINING LOOP
# -----------------------------

for epoch in range(num_epochs):

    for i, batch in enumerate(dataloader):

        # -----------------------------
        # Get real images
        # -----------------------------
        real_images = batch[0].to(device)

        # Create labels
        valid = torch.ones(real_images.size(0),1,device=device)
        fake = torch.zeros(real_images.size(0),1,device=device)


        # -----------------------------
        # TRAIN DISCRIMINATOR
        # -----------------------------

        optimizer_D.zero_grad()

        # Generate random noise
        z = torch.randn(real_images.size(0), latent_dim, device=device)

        # Generate fake images
        fake_images = generator(z)

        # Loss for real images
        real_loss = adversarial_loss(discriminator(real_images), valid)

        # Loss for fake images
        fake_loss = adversarial_loss(discriminator(fake_images.detach()), fake)

        # Total discriminator loss
        d_loss = (real_loss + fake_loss) / 2

        # Backpropagation
        d_loss.backward()
        optimizer_D.step()


        # -----------------------------
        # TRAIN GENERATOR
        # -----------------------------

        optimizer_G.zero_grad()

        gen_images = generator(z)

        # Generator tries to fool discriminator
        g_loss = adversarial_loss(discriminator(gen_images), valid)

        g_loss.backward()
        optimizer_G.step()


        # Print training progress
        if (i+1) % 100 == 0:
            print(
                f"Epoch [{epoch+1}/{num_epochs}] "
                f"Batch {i+1}/{len(dataloader)} "
                f"D Loss: {d_loss.item():.4f} "
                f"G Loss: {g_loss.item():.4f}"
            )


    # -----------------------------
    # Generate sample images
    # -----------------------------

    with torch.no_grad():
        z = torch.randn(16,latent_dim,device=device)

        generated = generator(z).detach().cpu()

        grid = torchvision.utils.make_grid(generated,nrow=4,normalize=True)

        plt.imshow(np.transpose(grid,(1,2,0)))
        plt.axis("off")
        plt.show()
```
